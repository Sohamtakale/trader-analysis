```python
trades['Timestamp IST'] = pd.to_datetime(
    trades['Timestamp IST'],
    format='mixed'
)
```

```python
trades['date'] = trades['Timestamp IST'].dt.date
trades['date'] = pd.to_datetime(trades['date'])
```

```python
sentiment['date'] = pd.to_datetime(sentiment['date'])
```

```python
merged = trades.merge(
    sentiment[['date', 'classification']],
    on='date',
    how='left'
)
```

```python
merged.head()
merged['classification'].isnull().sum()
```

```python
merged.columns
```

```python
daily_pnl = merged.groupby(['Account', 'date'])['Closed PnL'].sum().reset_index()
daily_pnl.rename(columns={'Closed PnL': 'daily_pnl'}, inplace=True)

daily_pnl.head()
```

```python
daily_trades = merged.groupby(['Account', 'date']).size().reset_index(name='trades_per_day')

daily_trades.head()
```

```python
merged['is_win'] = merged['Closed PnL'] > 0
```

```python
daily_winrate = merged.groupby(['Account', 'date'])['is_win'].mean().reset_index()
daily_winrate.rename(columns={'is_win': 'win_rate'}, inplace=True)

daily_winrate.head()
```

```python
daily_size = merged.groupby(['Account', 'date'])['Size USD'].mean().reset_index()
daily_size.rename(columns={'Size USD': 'avg_trade_size'}, inplace=True)

daily_size.head()
```

```python
merged['is_long'] = merged['Side'].str.upper() == 'BUY'
```

```python
daily_long_ratio = merged.groupby(['Account', 'date'])['is_long'].mean().reset_index()
daily_long_ratio.rename(columns={'is_long': 'long_ratio'}, inplace=True)

daily_long_ratio.head()
```

```python
daily_data = daily_pnl.merge(daily_trades, on=['Account', 'date'])
daily_data = daily_data.merge(daily_winrate, on=['Account', 'date'])
daily_data = daily_data.merge(daily_size, on=['Account', 'date'])
daily_data = daily_data.merge(daily_long_ratio, on=['Account', 'date'])
```

```python
daily_data = daily_data.merge(
    merged[['Account', 'date', 'classification']].drop_duplicates(),
    on=['Account', 'date'],
    how='left'
)
```

```python
daily_data.head()
daily_data.info()
```

```python
daily_data = daily_data.dropna(subset=['classification'])
```

```python
daily_data.info()
```

```python
daily_data['classification'] = daily_data['classification'].replace({
    'Extreme Fear': 'Fear',
    'Extreme Greed': 'Greed'
})
```
